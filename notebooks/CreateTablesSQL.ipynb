```python
import pandas as pd
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine, text
```

**Conexão com o BD**

```python
load_dotenv()  # Carrega variáveis do .env

endpoint = os.getenv("DB_ENDPOINT")
database = os.getenv("DB_NAME")
username = os.getenv("DB_USERNAME")
password = os.getenv("DB_PASSWORD")

connection_string = (
    f"mssql+pyodbc://{username}:{password}@{endpoint}:1433/{database}"
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&TrustServerCertificate=yes"
)
engine = create_engine(connection_string, fast_executemany=True)
```

**CREATE TABLE das tabelas do projeto**

```python
ddl = """
CREATE TABLE dbo.fato_vendas (
    id_venda   INT           PRIMARY KEY,
    dtVenda    DATE          NOT NULL,
    cdVendedor INT           NOT NULL,
    cdProduto  INT           NOT NULL,
    cdCliente  INT           NOT NULL,
    valor      DECIMAL(10,2) NOT NULL
);
"""

try: 
    with engine.begin() as conn:
        conn.execute(text(ddl))
except Exception as e:
    print("Erro ao criar a tabela fato_vendas: \n", e)
```

```python
ddl = """
CREATE TABLE dbo.clientes (
    cdCliente  INT            PRIMARY KEY,
    Nome       VARCHAR(100)   NOT NULL,
    Pais       VARCHAR(50)    NOT NULL,
    Regiao     VARCHAR(50)    NOT NULL,
    Cidade     VARCHAR(50)    NOT NULL,
    Endereco   VARCHAR(150)   NOT NULL,
    Email      VARCHAR(100)   NOT NULL
);
"""

try:
    with engine.begin() as conn:
        conn.execute(text(ddl))
except Exception as e:
    print("Erro ao criar a tabela clientes: \n", e)
```

```python
ddl = """
CREATE TABLE dbo.produtos (
    cdProduto  INT   PRIMARY KEY,
    descProduto      VARCHAR(100)   NOT NULL,
    qtdEstoque       INT    NOT NULL
);
"""

try: 
    with engine.begin() as conn:
        conn.execute(text(ddl))
except Exception as e:
    print("Erro ao criar a tabela produtos: \n", e)
```

```python
ddl = """
CREATE TABLE dbo.vendedores (
    cdVendedor  INT   PRIMARY KEY,
    nome        VARCHAR(100)   NOT NULL
);
"""

try: 
    with engine.begin() as conn:
        conn.execute(text(ddl))
except Exception as e:
    print("Erro ao criar a tabela vendedores: \n", e)
```
